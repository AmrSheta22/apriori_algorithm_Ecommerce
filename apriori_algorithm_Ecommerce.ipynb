```python
import pandas as pd
import numpy as np
from mlxtend.frequent_patterns import apriori
from mlxtend.frequent_patterns import association_rules
```

```python
df = pd.read_excel(r"C:\Users\amrsh\Downloads\archive(1)\US  E-commerce records 2020.xlsx")
df.head()
```

```python
data = df.copy()
```

```python
data.info()
```

```python
data.dropna(axis = 0, subset =['Order ID'], inplace = True)
tes = data.groupby('Order ID')
```

```python
# the three filters
data = data[data["Segment"] == "Consumer"]
data = data[data["Country"] == "United States"]
data = data[data["Profit"] > 1]
```

```python
descrip = {key: [] for key in data["Product Name"].unique()}
```

```python
droped = data.drop_duplicates("Order ID")
count = 0
for i in droped["Order ID"]:
    for j in tes.get_group(i)["Product Name"]:
        descrip[j].append(True)
    count +=1
    for v, k in descrip.items():
        if len(k)<count:
            descrip[v].append(False)
```

```python
dp = pd.DataFrame(descrip)
```

```python
dp.head(5)
```

```python
apri = apriori(dp, min_support=0.001, use_colnames=True)
apri
```

```python
rules = association_rules(apri, metric="lift", min_threshold=1)
rules = rules.sort_values(['confidence', 'lift'], ascending=[False, False])
rules.head()
```

```python
rules[(rules['lift'] >= 5) & (rules['confidence'] >= 0.4)].head()
```
